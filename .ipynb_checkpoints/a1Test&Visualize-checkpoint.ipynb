# Testing (Step 4 of Implementation Task)

```python
# Open all the images here
image1 = Image.open("./A1_Images/image1.jpg")
image2 = Image.open("./A1_Images/image2.jpg")
image3 = Image.open("./A1_Images/jiufen.jpeg")
```

```python
# show image 2 
plt.imshow(np.array(image2))
```

```python
# show image 3 (self selected image)
plt.imshow(np.array(image3))
```

```python
# Convert images into numpy arrays and convert them into gray scale
image_1 = np.array(image1)
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)
image_2 = np.array(image2)
image_2 = cv2.cvtColor(np.array(image2), cv2.COLOR_BGR2GRAY)
image_3 = np.array(image3)
image_3 = cv2.cvtColor(np.array(image3), cv2.COLOR_BGR2GRAY)
```

```python
# Running different steps of the edge detector for image 1

# Obtain gaussian matrix of kernel size = 3 and sigma = 0.5
gauss_matrix = step_1_get_gaussian_matrix(3, 0.5)
# Apply the gaussian filter on top of image_1 to get the blurred version of image_1
blurred_1 = convolve(gauss_matrix, image_1)
# Save the blurred image_1
plt.imsave("./output_images/blurred_1_sigma=0.5.jpg", blurred_1, cmap="gray")
# Computes the gradient magnitude matrix for image_1
gradient_1 = step_2_compute_gradient_magnitude(image_1)
# Save the gradient magnitude image for image_1
plt.imsave("./output_images/gradient_1_sigma=0.5.jpg", gradient_1, cmap="gray")
# Apply threshold onto the gradient magnitude matrix 
threshold_1 = step_3_threshold(gradient_1)
# Save the threshold image for image_1
plt.imsave("./output_images/threshold_1_sigma=0.5.jpg", threshold_1, cmap="gray")
# Show the threshold image for image_1
plt.imshow(threshold_1, cmap="gray")
```

```python
# Running different steps of the edge detector for image 2

# Obtain gaussian matrix of kernel size = 3 and sigma = 0.5
gauss_matrix = step_1_get_gaussian_matrix(3, 0.5)
# Apply the gaussian filter on top of image_2 to get the blurred version of image_2
blurred_2 = convolve(gauss_matrix, image_2)
# Save the blurred image_1
plt.imsave("./output_images/blurred_2_sigma=0.5.jpg", blurred_2, cmap="gray")
# Computes the gradient magnitude matrix for image_2
gradient_2 = step_2_compute_gradient_magnitude(image_2)
# Save the gradient magnitude image for image_2
plt.imsave("./output_images/gradient_2_sigma=0.5.jpg", gradient_2, cmap="gray")
# Apply threshold onto the gradient magnitude matrix
threshold_2 = step_3_threshold(gradient_2)
# Save the threshold image for image_2
plt.imsave("./output_images/threshold_2_sigma=0.5.jpg", threshold_2, cmap="gray")
# Show the threshold image for image_2
plt.imshow(threshold_2, cmap="gray")
```

```python
# Running different steps of the edge detector for image 3

# Obtain gaussian matrix of kernel size = 3 and sigma = 0.5
gauss_matrix = step_1_get_gaussian_matrix(3, 0.5)
# Apply the gaussian filter on top of image_2 to get the blurred version of image_2
blurred_3 = convolve(gauss_matrix, image_3)
# Save the blurred image_1
plt.imsave("./output_images/blurred_3_sigma=0.5.jpg", blurred_3, cmap="gray")
# Computes the gradient magnitude matrix for image_2
gradient_3 = step_2_compute_gradient_magnitude(image_3)
# Save the gradient magnitude image for image_2
plt.imsave("./output_images/gradient_3_sigma=0.5.jpg", gradient_3, cmap="gray")
# Apply threshold onto the gradient magnitude matrix
threshold_3 = step_3_threshold(gradient_3)
# Save the threshold image for image_2
plt.imsave("./output_images/threshold_3_sigma=0.5.jpg", threshold_3, cmap="gray")
# Show the threshold image for image_2
plt.imshow(threshold_3, cmap="gray")
```

## Template Code chunk for running edge detector on new image

```python
# Template for running edge_detector on new image
# //TODO: Add image name of newly added image
name = ""
new_image = Image.open(f"./A1_Images/{name}")
new_image = np.array(new_image)
new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2GRAY)
# Running different steps of the edge detector for new_image

# Obtain gaussian matrix of kernel size = 3 and sigma = 0.5
gauss_matrix = step_1_get_gaussian_matrix(3, 0.5)
# Apply the gaussian filter on top of image_3 to get the blurred version of new_image
blurred = convolve(gauss_matrix, new_image)
# Save the blurred new_image
plt.imsave("./output_images/blurred_new_sigma=0.5.jpg", blurred, cmap="gray")

# You can alter the kernel size and the sigma of the gaussian blur filter here
# Note that ksize MUST be odd
ksize = 3
scale = 0.5
# Computes the gradient magnitude matrix for new_image
gradient = step_2_compute_gradient_magnitude(new_image, ksize=ksize, scale=scale)
# Save the gradient magnitude image for new_image
plt.imsave("./output_images/gradient_new_sigma=0.5.jpg", gradient, cmap="gray")
# Apply threshold onto the gradient magnitude matrix
threshold = step_3_threshold(gradient)
# Save the threshold image for new_image
plt.imsave("./output_images/threshold_new_sigma=0.5.jpg", threshold, cmap="gray")
# Show the threshold image for new_image
plt.imshow(threshold, cmap="gray")
```
